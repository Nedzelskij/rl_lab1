```python
# !pip install gymnasium
import numpy as np
import gymnasium as gym
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. –û–∑–Ω–∞–π–æ–º–∏—Ç–∏—Å—å –∑ –æ–ø–∏—Å–æ–º —Å–µ—Ä–µ–¥–æ–≤–∏—â–∞ Frozen Lake –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ Gymnasium.

```python
# –°—Ç–≤–æ—Ä—é—î–º–æ —Å–µ—Ä–µ–¥–æ–≤–∏—â–µ Frozen Lake
env = gym.make('FrozenLake-v1', desc=["SFFF", "FHFH", "FFFH", "HFFG"], is_slippery=True)
env = env.unwrapped
desc=["SFFF", "FHFH", "FFFH", "HFFG"]
rows = len(desc)
cols = len(desc[0])
```

# 2. –û–±—á–∏—Å–ª–∏—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—é —Ü—ñ–Ω–∏ —Å—Ç–∞–Ω—É ùë£ùúã1(ùë†) –¥–ª—è —Ä—ñ–≤–Ω–æ–π–º–æ–≤—ñ—Ä–Ω–æ—ó (–≤–∏–ø–∞–¥–∫–æ–≤–æ—ó) —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó ùúã1 –ø—Ä–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä—ñ ùõæ = 0,5:

1.   –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —ñ—Ç–µ—Ä–∞—Ü—ñ–π–Ω–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É –û—Ü—ñ–Ω—é–≤–∞–Ω–Ω—è —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó (Policy Evaluation).
2.   –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ä–æ–∑–≤‚Äô—è–∑–∞–Ω–Ω—è —Å–∏—Å—Ç–µ–º–∏ —Ä—ñ–≤–Ω—è–Ω—å –ë–µ–ª–º–∞–Ω–∞ –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ—ó
—Ü—ñ–Ω–∏ —Å—Ç–∞–Ω—É –≤—ñ–¥–Ω–æ—Å–Ω–æ –Ω–µ–≤—ñ–¥–æ–º–∏—Ö –∑–Ω–∞—á–µ–Ω—å ùë•ùëñ = ùë£ùúã1(ùë†ùëñ)

–í–∏–≤–µ—Å—Ç–∏ –æ—Ç—Ä–∏–º–∞–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è —É –≤–∏–≥–ª—è–¥—ñ –º–∞—Ç—Ä–∏—Ü—ñ –∞–±–æ —Ç–µ–ø–ª–æ–≤–æ—ó –∫–∞—Ä—Ç–∏. –ß–∏
–±–∞—á–∏—Ç–µ –í–∏ –º–æ–∂–ª–∏–≤—ñ —à–ª—è—Ö–∏ –¥–æ –ø–æ–∫—Ä–∞—â–µ–Ω–Ω—è —Ä—ñ–≤–Ω–æ–π–º–æ–≤—ñ—Ä–Ω–æ—ó —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó?


```python
gamma = 0.5  # –∫–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç –∑–Ω–∏–∂–∫–∏
theta = 1e-8  # –ø–æ—Ä—ñ–≥ –¥–ª—è –∑—É–ø–∏–Ω–∫–∏
V = np.zeros(env.observation_space.n)  # —Ñ—É–Ω–∫—Ü—ñ—è –≤–∞—Ä—Ç–æ—Å—Ç—ñ —Å—Ç–∞–Ω—É
policy = np.ones([env.observation_space.n, env.action_space.n]) / env.action_space.n  # —Ä—ñ–≤–Ω–æ–π–º–æ–≤—ñ—Ä–Ω–∞ —Å—Ç—Ä–∞—Ç–µ–≥—ñ—è

def policy_evaluation(policy, env, gamma, theta):
    V = np.zeros(env.observation_space.n)  # –ø–æ—á–∞—Ç–∫–æ–≤—ñ –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó –≤–∞—Ä—Ç–æ—Å—Ç—ñ —Å—Ç–∞–Ω—É
    while True:
        delta = 0
        for s in range(env.observation_space.n):
            v = 0
            for a, action_prob in enumerate(policy[s]):
                for prob, next_state, reward, done in env.P[s][a]:
                    v += action_prob * prob * (reward + gamma * V[next_state])
            delta = max(delta, np.abs(V[s] - v))
            V[s] = v
        if delta < theta:
            break
    return V

# –û–±—á–∏—Å–ª—é—î–º–æ —Ñ—É–Ω–∫—Ü—ñ—é –≤–∞—Ä—Ç–æ—Å—Ç—ñ —Å—Ç–∞–Ω—É
V = policy_evaluation(policy, env, gamma, theta)

# –í–∏–≤–µ–¥–µ–Ω–Ω—è —Ç–µ–ø–ª–æ–≤–æ—ó –∫–∞—Ä—Ç–∏
V_grid = V.reshape((rows, cols))  # –ø–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ —É –º–∞—Ç—Ä–∏—Ü—é 4x4
plt.figure(figsize=(6, 6))
sns.heatmap(V_grid, annot=True, cmap="coolwarm", cbar=True)
plt.title("–§—É–Ω–∫—Ü—ñ—è —Ü—ñ–Ω–∏ —Å—Ç–∞–Ω—É V –¥–ª—è —Ä—ñ–≤–Ω–æ–π–º–æ–≤—ñ—Ä–Ω–æ—ó —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó")
plt.show()
```

```python
def policy_evaluation_LA(policy, env, gamma, theta):
    n_states = env.observation_space.n
    A = np.zeros((n_states, n_states))
    b = np.zeros(n_states)
    for s in range(n_states):
        A[s, s] = 1
        for a, action_prob in enumerate(policy[s]):
            for prob, next_state, reward, done in env.P[s][a]:
                A[s][next_state] -= action_prob * prob * gamma
                b[s] += action_prob * prob * reward
    V = np.linalg.solve(A, b)
    return V

# –†–æ–∑–≤'—è–∑—É—î–º–æ —Å–∏—Å—Ç–µ–º—É –ª—ñ–Ω—ñ–π–Ω–∏—Ö —Ä—ñ–≤–Ω—è–Ω—å
V = policy_evaluation_LA(policy, env, gamma, theta)

# –í–∏–≤–µ–¥–µ–Ω–Ω—è —Ç–µ–ø–ª–æ–≤–æ—ó –∫–∞—Ä—Ç–∏
V_grid = V.reshape((rows, cols))  # –ø–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ —É –º–∞—Ç—Ä–∏—Ü—é 4x4
plt.figure(figsize=(6, 6))
sns.heatmap(V_grid, annot=True, cmap="coolwarm", cbar=True)
plt.title("–§—É–Ω–∫—Ü—ñ—è —Ü—ñ–Ω–∏ —Å—Ç–∞–Ω—É V –¥–ª—è —Ä—ñ–≤–Ω–æ–π–º–æ–≤—ñ—Ä–Ω–æ—ó —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó")
plt.show()
```

# 3. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –∑–Ω–∞–π–¥–µ–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó —Ü—ñ–Ω–∏ —Å—Ç–∞–Ω—É ùë£ùúã1(ùë†) —Ç–∞ —Ä—ñ–≤–Ω—è–Ω–Ω—è –ë–µ–ª–º–∞–Ω–∞ –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ—ó —Ü—ñ–Ω–∏ –¥—ñ—ó-—Å—Ç–∞–Ω—É, –æ—Ü—ñ–Ω–∏—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—é —Ü—ñ–Ω–∏ –¥—ñ—ó-—Å—Ç–∞–Ω—É ùëûùúã1(ùë†ùëñ, ùëéùëó).


```python
def compute_q_value_function(V, env, gamma):
    Q = np.zeros((env.observation_space.n, env.action_space.n))
    for s in range(env.observation_space.n):
        for a in range(env.action_space.n):
            for prob, next_state, reward, done in env.P[s][a]:
                Q[s, a] += prob * (reward + gamma * V[next_state])
    return Q

# def q_from_v(P, V, s, gamma=1):
#     nS = V.shape[0]
#     nA = 4
#     q = np.zeros(nA)
#     for a in range(nA):
#         for prob, next_state, reward, done in P[s][a]:
#             q[a] += prob * (reward + gamma * V[next_state])
#     return q

# def Q_from_V(P, V, gamma=1):
#     nS = V.shape[0]
#     nA = 4
#     Q = np.zeros([nS, nA])
#     for s in range(nS):
#         Q[s] = q_from_v(P, V, s, gamma=gamma)
#     return Q

# –û–±—á–∏—Å–ª—é—î–º–æ —Ñ—É–Ω–∫—Ü—ñ—é —Ü—ñ–Ω–∏ –¥—ñ—ó-—Å—Ç–∞–Ω—É q(s, a)
Q = compute_q_value_function(V, env, gamma)
Q_equal = np.copy(Q)

# –í–∏–≤–µ–¥–µ–Ω–Ω—è —Ç–µ–ø–ª–æ–≤–æ—ó –∫–∞—Ä—Ç–∏
Q_grid = Q_equal.reshape((env.observation_space.n, env.action_space.n))  # –ø–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ —É –º–∞—Ç—Ä–∏—Ü—é 16x4
plt.figure(figsize=(6, 12))
sns.heatmap(Q_grid, annot=True, cmap="coolwarm", cbar=True)
plt.title("–§—É–Ω–∫—Ü—ñ—è —Ü—ñ–Ω–∏ –¥—ñ—ó-—Å—Ç–∞–Ω—É Q –¥–ª—è —Ä—ñ–≤–Ω–æ–π–º–æ–≤—ñ—Ä–Ω–æ—ó —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó")
plt.show()
```

# 4. –°—Ç–≤–æ—Ä–∏—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—é equiprobable, —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–º —è–∫–æ—ó —î –Ω–æ–º–µ—Ä –¥—ñ—ó. –î—ñ—è –æ–±–∏—Ä–∞—î—Ç—å—Å—è –≤–∏–ø–∞–¥–∫–æ–≤–∏–º —á–∏–Ω–æ–º –∑ –º–Ω–æ–∂–∏–Ω–∏ –¥–æ–ø—É—Å—Ç–∏–º–∏—Ö –¥—ñ–π.

```python
def equiprobable(env):
    action = np.random.choice(env.action_space.n)
    return action
```

# 5. –°—Ç–≤–æ—Ä–∏—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—é get_episode, —è–∫–∞ –ø—Ä–∏–π–º–∞—î —É —è–∫–æ—Å—Ç—ñ –∞—Ä–≥—É–º–µ–Ω—Ç—É –µ–∫–∑–µ–º–ø–ª—è—Ä —Å–µ—Ä–µ–¥–æ–≤–∏—â–∞, –∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–º —Ñ—É–Ω–∫—Ü—ñ—ó —î –µ–ø—ñ–∑–æ–¥, —Ç–æ–±—Ç–æ —Å–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂—ñ–≤, –∫–æ–∂–µ–Ω –∑ —è–∫–∏—Ö –∑–±–µ—Ä—ñ–≥–∞—î –≤—Å—ñ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ –∫–æ–∂–Ω–æ–≥–æ –∫—Ä–æ–∫—É –∞–≥–µ–Ω—Ç–∞ (—Ç–æ–±—Ç–æ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ–π —Å—Ç–∞–Ω, –¥—ñ—é, –≤–∏–Ω–∞–≥–æ—Ä–æ–¥—É, –ø–æ—Ç–æ—á–Ω–∏–π —Å—Ç–∞–Ω, –∑–Ω–∞—á–µ–Ω–Ω—è –ø–∞—Ä–∞–º–µ—Ç—Ä—ñ–≤ terminated —Ç–∞ truncated).
# –í–∏–±—ñ—Ä –∞–≥–µ–Ω—Ç–æ–º –¥—ñ—ó —É –∫–æ–∂–Ω–æ–º—É —Å—Ç–∞–Ω—ñ –Ω–∞ –¥–∞–Ω–æ–º—É –µ—Ç–∞–ø—ñ —Ä–µ–∞–ª—ñ–∑—É–π—Ç–µ –Ω–∞ –æ—Å–Ω–æ–≤—ñ —Ñ—É–Ω–∫—Ü—ñ—ó equiprobable, –∞–±–æ, —ñ–Ω—à–∏–º–∏ —Å–ª–æ–≤–∞–º–∏, –Ω–∞ –æ—Å–Ω–æ–≤—ñ —Ä—ñ–≤–Ω–æ–π–º–æ–≤—ñ—Ä–Ω–æ—ó (–≤–∏–ø–∞–¥–∫–æ–≤–æ—ó) —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó ùúã1.

```python
def get_episode(env):
    episode = []
    state, info = env.reset()  # –ü–æ—á–∞—Ç–∫–æ–≤–∏–π —Å—Ç–∞–Ω

    while True:
        action = equiprobable(env)  # –í–∏–±—ñ—Ä –≤–∏–ø–∞–¥–∫–æ–≤–æ—ó –¥—ñ—ó
        next_state, reward, terminated, truncated, info = env.step(action)  # –í–∏–∫–æ–Ω–∞–Ω–Ω—è –¥—ñ—ó

        episode.append((state, action, reward, next_state, terminated, truncated))

        state = next_state
        if terminated or truncated:
            break

    return episode
```

# 6. –í–∏–∫–æ–Ω–∞—Ç–∏ 75 –µ–ø—ñ–∑–æ–¥—ñ–≤ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ñ—É–Ω–∫—Ü—ñ—ó get_episode. –í–∏–≤–µ–¥—ñ—Ç—å –Ω–∞ –µ–∫—Ä–∞–Ω –¥–≤–∞ –≥—Ä–∞—Ñ—ñ–∫–∏: –≤–∏–Ω–∞–≥–æ—Ä–æ–¥–∞ —Ç–∞ —Ç—Ä–∏–≤–∞–ª—ñ—Å—Ç—å –µ–ø—ñ–∑–æ–¥—É.

```python
n_episodes = 75
rewards_equal = []
lengths_equal = []

for _ in range(n_episodes):
    episode = get_episode(env)

    total_reward = sum([step[2] for step in episode])
    rewards_equal.append(total_reward)

    episode_length = len(episode)
    lengths_equal.append(episode_length)

# –ì—Ä–∞—Ñ—ñ–∫ –≤–∏–Ω–∞–≥–æ—Ä–æ–¥
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(rewards_equal)
plt.title("–í–∏–Ω–∞–≥–æ—Ä–æ–¥–∞ –∑–∞ –µ–ø—ñ–∑–æ–¥")
plt.xlabel("–ù–æ–º–µ—Ä –µ–ø—ñ–∑–æ–¥—É")
plt.ylabel("–ó–∞–≥–∞–ª—å–Ω–∞ –≤–∏–Ω–∞–≥–æ—Ä–æ–¥–∞")

# –ì—Ä–∞—Ñ—ñ–∫ —Ç—Ä–∏–≤–∞–ª–æ—Å—Ç—ñ –µ–ø—ñ–∑–æ–¥—ñ–≤
plt.subplot(1, 2, 2)
plt.plot(lengths_equal)
plt.title("–¢—Ä–∏–≤–∞–ª—ñ—Å—Ç—å –µ–ø—ñ–∑–æ–¥—É")
plt.xlabel("–ù–æ–º–µ—Ä –µ–ø—ñ–∑–æ–¥—É")
plt.ylabel("–ö—ñ–ª—å–∫—ñ—Å—Ç—å –∫—Ä–æ–∫—ñ–≤")

plt.show()
```

# 7. –†–µ–∞–ª—ñ–∑—É–≤–∞—Ç–∏ –º–µ—Ç–æ–¥ –Ü—Ç–µ—Ä–∞—Ü—ñ—ó —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó (Policy Iteration) –¥–ª—è –∑–Ω–∞—Ö–æ–¥–∂–µ–Ω–Ω—è –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ—ó —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó –∑–∞ –∑–∞–¥–∞–Ω–æ—é –ø–æ—á–∞—Ç–∫–æ–≤–æ—é.

```python
def policy_improvement(env, policy, V, gamma, theta):
    policy_stable = True
    new_policy = np.zeros([env.observation_space.n, env.action_space.n])

    for s in range(env.observation_space.n):
        old_action = np.argmax(policy[s])
        action_values = np.zeros(env.action_space.n)

        # –û–±—á–∏—Å–ª—é—î–º–æ –∑–Ω–∞—á–µ–Ω–Ω—è –¥–ª—è –∫–æ–∂–Ω–æ—ó –¥—ñ—ó
        for a in range(env.action_space.n):
            for prob, next_state, reward, done in env.P[s][a]:
                action_values[a] += prob * (reward + gamma * V[next_state])

        # –í–∏–±–∏—Ä–∞—î–º–æ –≤—Å—ñ –¥—ñ—ó, —è–∫—ñ –º–∞—é—Ç—å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–µ –∑–Ω–∞—á–µ–Ω–Ω—è
        best_actions = np.flatnonzero(action_values == np.max(action_values))

        # –†—ñ–≤–Ω–æ–π–º–æ–≤—ñ—Ä–Ω–æ —Ä–æ–∑–ø–æ–¥—ñ–ª—è—î–º–æ –π–º–æ–≤—ñ—Ä–Ω–æ—Å—Ç—ñ –º—ñ–∂ –Ω–∞–π–∫—Ä–∞—â–∏–º–∏ –¥—ñ—è–º–∏
        new_policy[s, best_actions] = 1 / len(best_actions)

        if old_action != np.argmax(new_policy[s]):
            policy_stable = False

    return new_policy, policy_stable


def policy_iteration(env, policy, gamma, theta):
    V = np.zeros(env.observation_space.n)
    while True:
        V = policy_evaluation(policy, env, gamma, theta)
        policy, policy_stable = policy_improvement(env, policy, V, gamma, theta)

        if policy_stable:
            break

    return policy, V
```

# 8. –û—Ü—ñ–Ω–∏—Ç–∏ –æ–ø—Ç–∏–º–∞–ª—å–Ω—É —Å—Ç—Ä–∞—Ç–µ–≥—ñ—é ùúã‚àó —Ç–∞ —Ñ—É–Ω–∫—Ü—ñ—é —Ü—ñ–Ω–∏ —Å—Ç–∞–Ω—É ùë£_‚àó(ùë†) –¥–ª—è –∑–∞–¥–∞–Ω–æ–≥–æ —Å–µ—Ä–µ–¥–æ–≤–∏—â–∞ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –º–µ—Ç–æ–¥—É –Ü—Ç–µ—Ä–∞—Ü—ñ—ó —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ –≤ —è–∫–æ—Å—Ç—ñ –ø–æ—á–∞—Ç–∫–æ–≤–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä—ñ–≤ —Å—Ç—Ä–∞—Ç–µ–≥—ñ—é ùúã1 —Ç–∞ —Ñ—É–Ω–∫—Ü—ñ—é —Ü—ñ–Ω–∏ ùë£_ùúã1(ùë†). –ü–æ—Ä—ñ–≤–Ω—è–π—Ç–µ –æ—Ç—Ä–∏–º–∞–Ω—É —Ñ—É–Ω–∫—Ü—ñ—é —Ü—ñ–Ω–∏ ùë£_‚àó(ùë†) –∑ —Ñ—É–Ω–∫—Ü—ñ—î—é ùë£_ùúã1(ùë†) –∑ –∑–∞–≤–¥–∞–Ω–Ω—è ‚Ññ2.

```python
new_policy, new_V = policy_iteration(env, policy, gamma, theta)

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

V_grid_1 = V.reshape((rows, cols))
sns.heatmap(V_grid_1, annot=True, cmap="coolwarm", cbar=True, ax=axs[0])
axs[0].set_title("–§—É–Ω–∫—Ü—ñ—è —Ü—ñ–Ω–∏ —Å—Ç–∞–Ω—É V –¥–ª—è —Ä—ñ–≤–Ω–æ–π–º–æ–≤—ñ—Ä–Ω–æ—ó —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó")

V_grid_2 = new_V.reshape((rows, cols))
sns.heatmap(V_grid_2, annot=True, cmap="coolwarm", cbar=True, ax=axs[1])
axs[1].set_title("–û–ø—Ç–∏–º–∞–ª—å–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è —Ü—ñ–Ω–∏ —Å—Ç–∞–Ω—É V*")

# –í–∏–≤–µ–¥–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—ñ–≤
plt.tight_layout()
plt.show()

print("–û–ø—Ç–∏–º–∞–ª—å–Ω–∞ —Å—Ç—Ä–∞—Ç–µ–≥—ñ—è œÄ*:")
print(new_policy)
```

# 9. –í–∏–∫–æ–Ω–∞—Ç–∏ 12 –µ–ø—ñ–∑–æ–¥—ñ–≤ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ñ—É–Ω–∫—Ü—ñ—ó get_episode –∑—ñ —Å—Ç—Ä–∞—Ç–µ–≥—ñ—î—é ùúã‚àó. –í–∏–≤–µ–¥—ñ—Ç—å –Ω–∞ –µ–∫—Ä–∞–Ω –¥–≤–∞ –≥—Ä–∞—Ñ—ñ–∫–∏: –≤–∏–Ω–∞–≥–æ—Ä–æ–¥–∞ —Ç–∞ —Ç—Ä–∏–≤–∞–ª—ñ—Å—Ç—å –µ–ø—ñ–∑–æ–¥—É. –ü–æ—Ä—ñ–≤–Ω—è–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏ –∑ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–º–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏ –∑–∞–≤–¥–∞–Ω–Ω—è ‚Ññ6.


```python
def get_optimal_action(policy, state):
    action_probabilities = policy[state]
    optimal_action = action_probabilities.argmax()
    return optimal_action

def get_optimal_episode(env, policy):
    episode = []
    state, info = env.reset()  # –ü–æ—á–∞—Ç–∫–æ–≤–∏–π —Å—Ç–∞–Ω

    while True:
        action = get_optimal_action(policy, state)  # –í–∏–±—ñ—Ä –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ—ó –¥—ñ—ó
        next_state, reward, terminated, truncated, info = env.step(action)  # –í–∏–∫–æ–Ω–∞–Ω–Ω—è –¥—ñ—ó

        episode.append((state, action, reward, next_state, terminated, truncated))

        state = next_state
        if terminated or truncated:
            break

    return episode
```

# 11.–°—Ç–≤–æ—Ä–∏—Ç–∏ —Ñ—É–Ω–∫—Ü—ñ—é eps_greedy_policy, –∞—Ä–≥—É–º–µ–Ω—Ç–∞–º–∏ —è–∫–æ—ó —î –º–∞—Å–∏–≤ –∑–Ω–∞—á–µ–Ω—å —Ñ—É–Ω–∫—Ü—ñ—ó —Ü—ñ–Ω–∏ –¥—ñ—ó-—Å—Ç–∞–Ω—É ùëû(ùë†, ùëé) —Ç–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä ùúÄ, —Ç–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–º —î –Ω–æ–º–µ—Ä –¥—ñ—ó, –æ–±—Ä–∞–Ω–∏–π –∑ –º–Ω–æ–∂–∏–Ω–∏ –Ω–æ–º–µ—Ä—ñ–≤ –¥–æ–ø—É—Å—Ç–∏–º–∏—Ö –¥—ñ–π –¥–æ–ø–æ–º–æ–≥–æ—é –º–µ—Ç–æ–¥—É ùúÄ-–∂–∞–¥—ñ–±–Ω–æ—ó —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó (ùúÄ-greedy policy).

```python
def eps_greedy_policy(Q_values, epsilon):
    """
    –í–∏–±–∏—Ä–∞—î –¥—ñ—é –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é eps-–∂–∞–¥—ñ–±–Ω–æ—ó —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó.

    –ê—Ä–≥—É–º–µ–Ω—Ç–∏:
    Q_values -- –º–∞—Å–∏–≤ –∑–Ω–∞—á–µ–Ω—å q(s, a) –¥–ª—è –≤—Å—ñ—Ö –¥—ñ–π —É —Å—Ç–∞–Ω—ñ s.
    epsilon -- –ø–∞—Ä–∞–º–µ—Ç—Ä Œµ, —â–æ –≤–∏–∑–Ω–∞—á–∞—î –π–º–æ–≤—ñ—Ä–Ω—ñ—Å—Ç—å –≤–∏–ø–∞–¥–∫–æ–≤–æ–≥–æ –≤–∏–±–æ—Ä—É –¥—ñ—ó.

    –ü–æ–≤–µ—Ä—Ç–∞—î:
    –í–∏–±—Ä–∞–Ω–∏–π –Ω–æ–º–µ—Ä –¥—ñ—ó.
    """
    if np.random.rand() < epsilon:
        return np.random.choice(len(Q_values))
    else:
        return np.argmax(Q_values)
```

# 12.–†–µ–∞–ª—ñ–∑—É–≤–∞—Ç–∏ –º–µ—Ç–æ–¥ SARSAmax –¥–ª—è –∑–Ω–∞—Ö–æ–¥–∂–µ–Ω–Ω—è –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó —Ü—ñ–Ω–∏ –¥—ñ—ó-—Å—Ç–∞–Ω—É ùëû‚àó(ùë†, ùëé) –∑–∞ –∑–∞–¥–∞–Ω–æ—é –ø–æ—á–∞—Ç–∫–æ–≤–æ—é.


```python
def sarsa_max(env, Q, num_episodes, alpha, gamma, epsilon):
    """
    –†–µ–∞–ª—ñ–∑–∞—Ü—ñ—è –º–µ—Ç–æ–¥—É SARSAmax –¥–ª—è –ø–æ—à—É–∫—É –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó —Ü—ñ–Ω–∏ –¥—ñ—ó-—Å—Ç–∞–Ω—É Q_*(s, a).

    –ê—Ä–≥—É–º–µ–Ω—Ç–∏:
    env -- —Å–µ—Ä–µ–¥–æ–≤–∏—â–µ Gymnasium.
    num_episodes -- –∫—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ø—ñ–∑–æ–¥—ñ–≤ –¥–ª—è –Ω–∞–≤—á–∞–Ω–Ω—è.
    alpha -- —à–≤–∏–¥–∫—ñ—Å—Ç—å –Ω–∞–≤—á–∞–Ω–Ω—è.
    gamma -- –∫–æ–µ—Ñ—ñ—Ü—ñ—î–Ω—Ç –¥–∏—Å–∫–æ–Ω—Ç—É–≤–∞–Ω–Ω—è.
    epsilon -- –ø–∞—Ä–∞–º–µ—Ç—Ä Œµ –¥–ª—è eps-–∂–∞–¥—ñ–±–Ω–æ—ó —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó.

    –ü–æ–≤–µ—Ä—Ç–∞—î:
    –ú–∞—Å–∏–≤ –∑–Ω–∞—á–µ–Ω—å —Ñ—É–Ω–∫—Ü—ñ—ó Q(s, a).
    """
    Q_values = np.copy(Q)

    for episode in range(num_episodes):
        state, _ = env.reset()
        done = False

        while not done:
            # –û–±–∏—Ä–∞—î–º–æ –¥—ñ—é –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é eps-–∂–∞–¥—ñ–±–Ω–æ—ó —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó
            action = eps_greedy_policy(Q_values[state], epsilon)

            # –í–∏–∫–æ–Ω—É—î–º–æ –¥—ñ—é —Ç–∞ –æ—Ç—Ä–∏–º—É—î–º–æ –Ω–æ–≤–∏–π —Å—Ç–∞–Ω, –≤–∏–Ω–∞–≥–æ—Ä–æ–¥—É —Ç–∞ —Å—Ç–∞—Ç—É—Å –∑–∞–≤–µ—Ä—à–µ–Ω–Ω—è
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated

            # –û–Ω–æ–≤–ª—é—î–º–æ –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó q(s, a) –∑–∞ –º–µ—Ç–æ–¥–æ–º SARSAmax
            next_action = eps_greedy_policy(Q_values[next_state], epsilon)
            td_target = reward + gamma * np.max(Q_values[next_state])
            td_error = td_target - Q_values[state][action]
            Q_values[state][action] += alpha * td_error

            state = next_state

    return Q_values
```

# 13.–û—Ü—ñ–Ω–∏—Ç–∏ –æ–ø—Ç–∏–º–∞–ª—å–Ω—É —Ñ—É–Ω–∫—Ü—ñ—é —Ü—ñ–Ω–∏ –¥—ñ—ó-—Å—Ç–∞–Ω—É ùëû‚àó(ùë†, ùëé) –¥–ª—è –∑–∞–¥–∞–Ω–æ–≥–æ —Å–µ—Ä–µ–¥–æ–≤–∏—â–∞ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –º–µ—Ç–æ–¥—É SARSAmax, –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ ùúÄ-–∂–∞–¥—ñ–±–Ω—É —Å—Ç—Ä–∞—Ç–µ–≥—ñ—é —Ç–∞ —Ñ—É–Ω–∫—Ü—ñ—é —Ü—ñ–Ω–∏ –¥—ñ—ó-—Å—Ç–∞–Ω—É ùëûùúã1(ùë†, ùëé).
# –í–∏–∫–æ—Ä–∏—Å—Ç–∞–π—Ç–µ –∑–Ω–∞—á–µ–Ω–Ω—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ ùúÄ ‚àà {0,1, 0,5, 0,9}. –ü–æ—Ä—ñ–≤–Ω—è–π—Ç–µ –æ—Ç—Ä–∏–º–∞–Ω—ñ –æ—Ü—ñ–Ω–∫–∏ —Ñ—É–Ω–∫—Ü—ñ–π ùëû‚àó(ùë†, ùëé) –∑ —Ñ—É–Ω–∫—Ü—ñ—è–º–∏ ùëûùúã1(ùë†, ùëé) —Ç–∞ ùëû‚àó(ùë†, ùëé) –∑ –∑–∞–≤–¥–∞–Ω—å ‚Ññ‚Ññ3, 10.
# –û–±–µ—Ä—ñ—Ç—å –∫—Ä–∞—â—É –æ—Ü—ñ–Ω–∫—É —Ñ—É–Ω–∫—Ü—ñ—ó ùëû‚àó(ùë†, ùëé) —ñ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–µ —ó–π –∑–Ω–∞—á–µ–Ω–Ω—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ ùúÄ.

```python
alpha = 0.1  # –®–≤–∏–¥–∫—ñ—Å—Ç—å –Ω–∞–≤—á–∞–Ω–Ω—è
num_episodes = 5000  # –ö—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ø—ñ–∑–æ–¥—ñ–≤ –¥–ª—è –Ω–∞–≤—á–∞–Ω–Ω—è

# –û—Ü—ñ–Ω–∫–∞ –¥–ª—è epsilon = 0.1
epsilon = 0.1
Q_eps_01 = sarsa_max(env, Q, num_episodes, alpha, gamma, epsilon)

# –û—Ü—ñ–Ω–∫–∞ –¥–ª—è epsilon = 0.5
epsilon = 0.5
Q_eps_05 = sarsa_max(env, Q, num_episodes, alpha, gamma, epsilon)

# –û—Ü—ñ–Ω–∫–∞ –¥–ª—è epsilon = 0.9
epsilon = 0.9
Q_eps_09 = sarsa_max(env, Q, num_episodes, alpha, gamma, epsilon)

# –¢–µ–ø–ª–æ–≤—ñ –∫–∞—Ä—Ç–∏ –¥–ª—è –ø–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è q(s, a) –ø—Ä–∏ —Ä—ñ–∑–Ω–∏—Ö epsilon
plt.figure(figsize=(30, 10))

plt.subplot(1, 5, 1)
sns.heatmap(Q_eps_01, annot=True, cmap="coolwarm")
plt.title('Q –¥–ª—è epsilon = 0.1')

plt.subplot(1, 5, 2)
sns.heatmap(Q_eps_05, annot=True, cmap="coolwarm")
plt.title('Q –¥–ª—è epsilon = 0.5')

plt.subplot(1, 5, 3)
sns.heatmap(Q_eps_09, annot=True, cmap="coolwarm")
plt.title('Q –¥–ª—è epsilon = 0.9')

plt.subplot(1, 5, 4)
sns.heatmap(Q, annot=True, cmap="coolwarm")
plt.title('Q –¥–ª—è —Ä—ñ–≤–Ω–æ–º—ñ—Ä–Ω–æ—ó —Å—Ç—Ä–∞—Ç–µ–≥—ñ—ó')

plt.subplot(1, 5, 5)
sns.heatmap(new_Q, annot=True, cmap="coolwarm")
plt.title('Q*')

plt.show()
```

# 14. –°—Ç–≤–æ—Ä—ñ—Ç—å –Ω–∞ –æ—Å–Ω–æ–≤—ñ —Ñ—É–Ω–∫—Ü—ñ—ó ùëû‚àó(ùë†, ùëé) –∑ –∑–∞–≤–¥–∞–Ω–Ω—è ‚Ññ13 ùúÄ-–∂–∞–¥—ñ–±–Ω—É —Å—Ç—Ä–∞—Ç–µ–≥—ñ—é ùúã2.


```python
def create_eps_greedy_policy(Q, epsilon, env):
    policy = np.zeros([env.observation_space.n, env.action_space.n])
    for s in range(env.observation_space.n):
        Q_values = Q[s]
        if np.random.rand() < epsilon:
            best_actions = [np.random.choice(len(Q_values))]
        else:
            best_actions = np.flatnonzero(Q_values == np.max(Q_values))

        policy[s, best_actions] = 1 / len(best_actions)

    return policy

# –°—Ç–≤–æ—Ä—é—î–º–æ Œµ-–∂–∞–¥—ñ–±–Ω—É —Å—Ç—Ä–∞—Ç–µ–≥—ñ—é œÄ2 –¥–ª—è epsilon = 0.1 –Ω–∞ –æ—Å–Ω–æ–≤—ñ q*(s, a)
epsilon = 0.1
policy_sarsa_max = create_eps_greedy_policy(Q_eps_01, epsilon, env)

print("\n–û—Ü—ñ–Ω–µ–Ω–∞ –æ–ø—Ç–∏–º–∞–ª—å–Ω—É —Å—Ç—Ä–∞—Ç–µ–≥—ñ—è (–ª—ñ–≤–æ = 0, –Ω–∏–∑ = 1, –ø—Ä–∞–≤–æ—Ä—É—á = 2, –≤–≤–µ—Ä—Ö = 3):")
print(policy_sarsa_max)
```
